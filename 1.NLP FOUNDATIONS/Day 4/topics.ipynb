BOW -
A technique that converts text into numbers based on word frequencies, ignoring grammer or order

Simple Example

Suppose we have two sentences:

Sentence 1: "I love NLP"
Sentence 2: "I love Python and NLP"


Vocabulary (unique words from all sentences):

["I", "love", "NLP", "Python", "and"]


Now count frequency in each sentence:

Sentence I love NLP	Python and

S1	   --  1	 1	 1	  0     0

S2	  --   1	 1	 1	  1  	1

This matrix representation is Bag of Words.

Why is it called "Bag"?

Because if you scramble the sentence:

"NLP love I" = "I love NLP"

BoW treats them the same since only word counts matter.

Sparse Matrix & Vocabulary Explosion

A matrix is sparse if most of its values are 0

Example: If a vocabulary has 20,000 words, but a sentence contains only 10 words, then:

only 10 cells = non-zero
19,990 cells = zero

it will create a matrix for every word in the vocabulary which is like group of seperate words ... consider a sentance "Hey There, I'm MR.X(approx. 6 tokens)" ... and your vocabulary has 100 words in it ... it creates a 6 x 100 2d matrix ...if the word on the sentance os present in the vocabulary set it put 1 else it makes it zero

TF-IDF - Term frequency - inverse document frequency

A weighting method that reduces the importance of common words and increases the importance of rare but meaningful words.

TF-IDF highlights keywords and reduces noise like "the", "is", "and", "I".

FORMULA:

TF:
 How many times a word appears in a document
  
    Word count in document / Total words in document

IDF:
 Measures how rare a word is across document

        IDF = log(N/df)

        N -> total number of documents
        df -> # of documents containing the word

        Final TF-IDF Score:
            TF-IDF = TF * IDF
