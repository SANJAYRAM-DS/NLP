Text cleaning and REGEX

REGEX is a pattern-bassed language used to search, match, and replace text

REGEX works like smart filters. Insted of removing letters manually, you tell Python: Hey, remove anything that looks like an URL--

Some patterns

Match digits - \d

Match alphabets - [a-zA-Z]

Match space - \s

Match any characted - .

Match email - [\w\.-]+@[\w\.-]+
