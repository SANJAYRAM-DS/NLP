```python
import re
```

```python
#Finding numbers
text = "Hello1 world2 3 456"
pattern = r"\d+"

print(re.findall(pattern, text))
```

```python
#Removing URLs
text = "Check this link https://google.com for details"
clean = re.sub(r'http\S+|www\S+', '', text)
print(clean)
```

```python
#Removing Email Addresses
text = "Contact me at sanjayram.data@gmail.com"
clean = re.sub(r'\S+@\S+', '', text)
print(clean)
```

```python
#Removing mentions
text = "Hello @sanjayram, how are you?"
clean = re.sub(r'@\w+', '', text)
clean_ = re.sub(r'@', '', text)
print(clean)
print(clean_)
```

```python
#Removing numbers
text = "He1llo wo0rld"
clean = re.sub(r'\d', '', text)
print(clean)
```

```python
#Handiling emojies
text = "Handling Emojis ðŸ˜ŠðŸ”¥ðŸ’¯"
emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F"  # emoticons
                            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                            u"\U0001F680-\U0001F6FF"  # transport & map
                            u"\U0001F1E0-\U0001F1FF"  # flags
                            "]+", flags=re.UNICODE)

clean_text = emoji_pattern.sub(r'', text)
print(clean_text)
```

```python
#Handiling slang / Misspelling
slang_dict = {
    "u": "you",
    "r": "are",
    "idk": "I don't know",
    "lol": "laugh out loud",
}

def normalize_slang(text):
    words = text.split()
    return " ".join([slang_dict.get(w, w) for w in words])
```
