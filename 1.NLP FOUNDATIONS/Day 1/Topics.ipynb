Natural Language Processing (NLP) is a field of AI that enables machines to understand, interpret, generate, and interact using human languages (like English, Hindi, Tamil, etc.).

Computers don’t understand text naturally — they understand numbers. NLP converts language → structured data → numbers → decisions.

NLP - Natural Language Processing - the entire field

NLU - Natural Language Understanding - Machine understandinds Meaning -> "I hate this" -> Negative

NLG - Natural Language Generation - Machine creates Language ->Gpt, Test Summarization

Text Pipelines:
Raw Text -> Normalization -> Tokenization -> StopWord Removal -> Stemming / Lemmatization -> Vectorization

Syntax -> Grammar and Structure

Sematics -> Meaning of Sentance

Corpus -> Large dataset of text used for training NLP models like paragraphs

Annotations -> Addin labels like word(California) - Label(Location)

Metadata -> Extra information like language, author, timestamp


TOKEN -> Individual word after splitting -> eg: ["I'm loving it"] -> ["I","m","loving","it"]

TYPE -> Unique Token -> "I love I" -> Token: 3, TYPE: 2(I, love)

VOCABULARY -> Set of all unique words in dataset -> I, love

OOV(Out-Of-Vocabualry) -> Word model has never seen before -> "Zynaptic" if never in training

Raw text
  └─► Text normalization (encoding, unicode, case)
      └─► Sentence segmentation
          └─► Tokenization (word / subword / char)
              └─► Cleaning (punctuation, URLs, mentions)
                  └─► Contraction handling
                      └─► Stopword handling
                          └─► Spelling correction / normalization
                              └─► Lemmatization / Stemming
                                  └─► POS tagging / NER / parsing (optional features)
                                      └─► Vectorization (BoW / TF-IDF / embeddings)
